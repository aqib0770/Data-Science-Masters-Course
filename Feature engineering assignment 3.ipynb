Q1. What is data encoding? How is it useful in data science?
<br>Ans. Data coding is the process of converting categorical data into numerical data. When building Machine Learning models, the models don't understand categorical or non-numerical data, they only understand numerical data. So to build such Machine Leraning models, Data Encoding is used.

Q2. What is nominal encoding? Provide an example of how you would use it in a real-world scenario.
<br>Ans. When we convert a categorical data into numerical data where rank or order of category dosen't matter, then it is called nominal encoding.
<br>Lets suppose we have a categorical data of marital status of person 'married','unmarried','other' which can be encoded respectively with 3 different unique labels [0,1,2]

Q3. In what situations is nominal encoding preferred over one-hot encoding? Provide a practical example.
<br>Ans. When the number of categories in a feature is very large, we use nominal encoding because it assigns a unique label to each category while one-hot encoding can result in large number of binary columns.
<br>Lets suppose we have a feature which consists of various cities of India, in this situation one-hot encoding would create binary columns for each country, resulting in large number of columns. While nominal encoding would simply assign a unique labe to each unique city

Q4. Suppose you have a dataset containing categorical data with 5 unique values. Which encoding
technique would you use to transform this data into a format suitable for machine learning algorithms?
Explain why you made this choice.
<br>Ans. To transform a categorical data with 5 unique values into a format suitable for machine learning, we will use one-hot encoding because:
1. It creates binary columns for each category which allows models to work effectively.
2. One-hot encoding ensures that each category is represented independently as a separate binary column so that we can avoid any intended relationship between categories that might be falsely applied by a nominal encoding

Q5. In a machine learning project, you have a dataset with 1000 rows and 5 columns. Two of the columns
are categorical, and the remaining three columns are numerical. If you were to use nominal encoding to
transform the categorical data, how many new columns would be created? Show your calculations.

Ans. In nominal encoding, every unique value is replaced by a numerical value. So the original column would get replaced by a single encoded column consist of 1000 rows. Therefore, no new columns will be created.

## These are the calculations

```python
import pandas as pd
import numpy as np
```

```python
age=np.random.randint(18,70,size=1000)
weight=np.random.randint(20,80,size=1000)
height=np.random.randint(80,180,size=1000)

#Creating categories for 2 categorical columns

gender_cat=['male','female']
cities=['Mumbai','Delhi','Banglore','Hyderabad']

city=np.random.choice(cities,size=1000)
gender=np.random.choice(gender_cat,size=1000)
```

```python
#Creating a dataframe from above data
df=pd.DataFrame({'Age':age,'Wieght (in kg)':weight,'Height (in cm)':height,'Gender':gender,'City':city})
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder #Label encoding and nominal encoding are same
```

```python
encoder=LabelEncoder()
```

```python
df['City']=encoder.fit_transform(df[['City']])
df['Gender']=encoder.fit_transform(df[['Gender']])
```

Q7.You are working on a project that involves predicting customer churn for a telecommunications
company. You have a dataset with 5 features, including the customer's gender, age, contract type,
monthly charges, and tenure. Which encoding technique(s) would you use to transform the categorical
data into numerical data? Provide a step-by-step explanation of how you would implement the encoding.

Ans. If a am working on a dataset which consists of fetaures like gender,age,contract type,monthly charges and tenure then I will use only one-hot encoding for gender and contract type features because all others features are already numerical.

```python
age=np.random.randint(18,50,size=100)
gender=np.random.choice(['Male','female'],size=100)
contract_type=np.random.choice(['Monthly','Quaterly','Yearly'],size=100)
monthly_charges=np.random.randint(100,1000,size=100)
tenure=np.random.choice([6,12,18,24],size=100)
```

```python
data=pd.DataFrame({'Age (in years)':age,'Gender':gender,'Contract type':contract_type,'Monthly charges (in rupees)':monthly_charges,'Tenure (in months)':tenure})
```

```python
data
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encode=OneHotEncoder()
```

```python
encoded=encode.fit_transform(data[['Gender','Contract type']]).toarray()
```

```python
encoded_df=pd.DataFrame(encoded,columns=encode.get_feature_names_out())
```
