Q1. What is Bayes' theorem?
<br>Ans: Bayes' theorem is a mathematical formula used in probability and statistics. It describes the probability of an event, based on prior knowledge of conditions that might be related to the event.
<br>Formula: P(A/B)=(P(B/A)*P(A))/P(B)

Q2. What is the formula for Bayes' theorem?
<br>Ans: Formula: P(A/B)=(P(B/A)*P(A))/P(B)

Q3. How is Bayes' theorem used in practice?
- Medical Diagnosis: Bayes theorem is used in medical diagnosis, where the probability of a particular disease or condition given certain symptoms or test results is calculated.
- Machine learning: Bayes theorem finds wide application in applied machine learning and establishes a relationship between data and model.

Q4. What is the relationship between Bayes' theorem and conditional probability?
<br>Ans: Conditional probability is the probability of one event occuring given that another event has already occured.
<br>Bayes' theorem is the way of using conditional probability to update our beliefs about the probability of an event occuring, given new information

Q5. How do you choose which type of Naive Bayes classifier to use for any given problem?
- Multinomial naive bayes: When the features are discrete count, such as number of words, Multinomial naive bayes is used.
- Bernoulli naive bayes: When the feature values are binary, Bernoulli naive bayes is used.
- Gaussian naive bayes: When the feature values are continuous and normally distributed then Gaussioan naive bayes is used.
