Q2. Write a python program to print Python Exception Hierarchy.

```python
# Recursively prints the exception hierarchy
def print_exception_hierarchy(exc_class, indent=0):
    print(' ' * indent, exc_class.__name__)
    for sub_class in exc_class.__subclasses__():
        print_exception_hierarchy(sub_class, indent + 2)

# Start with the base Exception class
print_exception_hierarchy(BaseException, 0)
```

Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

Ans) FloatingPointError
       ,OverflowError
       ,ZeroDivisionError are the errors defined under ArithmeticError class.

```python
#zero division error
try:
    50/0
except ZeroDivisionError as e:
    print(e)
```

```python
a = 5.0

try:
    for i in range(1, 1000):
        a = a**i
except ArithmeticError as e:
    print(e)
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans) LookupError class is used to represent the errors that occurs while looking up an object in tuple, list or any data structure and that object is not found.

```python
#keyError
d={"name":"Aqib","surname":"Ansari"}
try:
    d["age"]
except KeyError as k:
    print("This code will raise a key error:",k)
```

```python
#index error
l=[1,2,3,["aqib","Ansari"]]
try:
    l[6]
except IndexError as i:
    print(i)
```

Q5. Explain ImportError. What is ModuleNotFoundError?

Ans) ImportError is an exception raised when an imported module or object cannot be found. It occurs when either module or object is not found or when the imported module or object has syntax errors.
<br/>ModuleNotFoundError is an exception raised when an imported module, or an object within an imported module is not found
