Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

Ans) A scatter plot is a plot which uses a dot symbol to show the data.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("This is a scatter plot")
```

The subplot() function is used to create multiple plots in a single figure

```python
x1=np.array([0, 1, 2, 3, 4, 5])
y1=np.array([0, 100, 200, 300, 400, 500])
```

```python
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
```

```python
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
```

```python
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(2,2,(1)) ## in this function the first and second argument (2,2) is the shape of the figure and third argument (1) is the position of the plot
plt.plot(x1,y1)
plt.title("Line plot 1")
plt.subplot(2,2,(2))
plt.plot(x2,y2)
plt.title("Line plot 2")
plt.subplot(2,2,(3))
plt.plot(x3,y3)
plt.title("Line plot 3")
plt.subplot(2,2,(4))
plt.plot(x4,y4)
plt.title("Line plot 4")
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot is a way to represent the data using rectangular bars. It is used to visually represent the data and compare the values of different categories.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(6,3))
plt.bar(company,profit)
plt.title("Vertical plot")
```

```python
plt.figure(figsize=(6,3))
plt.barh(company,profit)
plt.title("Horizontal plot")
```
