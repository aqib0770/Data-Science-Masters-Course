Q1. What is the difference between Ordinal Encoding and Label Encoding? Provide an example of when you
might choose one over the other.
<br>Ans. Ordinal encoding and Label encoding are the two techniques which convert categorical data into numerical data. Ordinal encoding assigns numercal values based on the rank or order of categories while Label encoding assigns values without any rank or order.
<br>Suppose there is a dataset of grades of students which consists of three categories [A,B,C]. In this cas we will choose ordinal encoding so that it assign values based on rank like A is greater rank and C is the smallest rank.

Q2. Explain how Target Guided Ordinal Encoding works and provide an example of when you might use it in
a machine learning project.
<br>Ans. In target guided ordinal encoding we replace the categorical feature with mean of the target variable of that category.

```python
import pandas as pd
import numpy as np
```

```python
players=np.random.choice(['Virat Kohli','Steve Smith','Kane Williamson'],size=100)
runs=np.random.randint(10,200,size=100)
```

```python
player_data=pd.DataFrame({'Players':players,'Runs':runs})
```

```python
average_score=player_data.groupby('Players')['Runs'].mean().to_dict()
```

```python
player_data['Average Score']=player_data['Players'].map(average_score)
```

Q3. Define covariance and explain why it is important in statistical analysis. How is covariance calculated?
<br>Ans. Covariance is the method of finding relationship between features of a dataset. 
1. Covariance helps in determining the direction of relationship. A positive covariance indicates direct relationship where both variables increase or decrease together. A negative covariance indicates inverse relationship where one variable is increasing and the other variable is decreasing.
2. Covariance helps in finding the strength of relationship between variables which can help in selecting features which are highly related
<br>Covariance calculation: Σ((xi-X̄)(yi-ȳ))/n-1

Q4. For a dataset with the following categorical variables: Color (red, green, blue), Size (small, medium,
large), and Material (wood, metal, plastic), perform label encoding using Python's scikit-learn library.
Show your code and explain the output.

```python
color=np.random.choice(['red','green','blue'],size=100)
size=np.random.choice(['small','medium','large'],size=100)
material=np.random.choice(['wood','metal','plastic'],size=100)

data=pd.DataFrame({'color':color,'size':size,'material':material})
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
data['Color encoded']=encoder.fit_transform(data['color'])
data['Size encoded']=encoder.fit_transform(data['size'])
data['Material encoded']=encoder.fit_transform(data['material'])
data
```

In output, we added 3 more features [Color encoded,Size encoded,Material encoded] which contain encoded value of features [Color,Size',Material] respectively

Q7. You are analyzing a dataset with two continuous variables, "Temperature" and "Humidity", and two
categorical variables, "Weather Condition" (Sunny/Cloudy/Rainy) and "Wind Direction" (North/South/
East/West). Calculate the covariance between each pair of variables and interpret the results.

```python
temperature=np.random.uniform(20,35,size=100) # Generating 100 samples of temperature between 20 degree celcius to 35 degree
humidity=np.random.uniform(40,80,size=100)# Generating samples of humidity between 40% to 80%
weather=np.random.choice(['Sunny','Cloudy','Rainy'],size=100)
wind=np.random.choice(['North','South','East','West'],size=100)
```

```python
atmosphere=pd.DataFrame({'Temperature':temperature,'Humidity':humidity,'Weather Condition':weather,'Wind Direction':wind})
atmosphere
```

```python
encoded_df = pd.get_dummies(atmosphere[['Weather Condition', 'Wind Direction']])
encoded_atmosphere=pd.concat([atmosphere,encoded_df],axis=1)
encoded_atmosphere
```

```python
encoded_atmosphere.cov()
```
