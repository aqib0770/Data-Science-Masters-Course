Q1. What is Flask Framework? What are the advantages of Flask Framework?

Ans) Flask is a micro web framework written in Python which is used for developing web applications. It is very light and flexible. Its advantages are:
1. It is easy to develop and maintain websites with Flask.
2. It has simple and easy URL routing.
3. It can scale quickly for growing web applications.

Q3.What is App routing in Flask? Why do we use app routes?

Ans) App routing is a technique to map a specific URL with a funvtion that handles it. We use app routes to help users remember application URLs and access the desired page directly.
